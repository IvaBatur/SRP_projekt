```python
import pandas as pd
import random

CSV_FILE_PATH = "hotel_bookings.csv"
df = pd.read_csv(CSV_FILE_PATH)
print("CSV size before: ", df.shape)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_') 
df = df.drop(columns=['company', 'agent'])
print("CSV size after:", df.shape )
```

```python
import pycountry
def fix_country_names(code):
    try:
        return pycountry.countries.get(alpha_3=code).name
    except:
        return code

df['country'] = df['country'].apply(fix_country_names)
df['country'] = df['country'].replace({'TMP':'East Timor',})
```

```python
import random
def randomize_location(df, column_name, country_list):

    num_to_change = int(len(df) * 0.6)

    indices = df.index.tolist()
    rows_to_change = random.sample(indices, min(num_to_change, len(indices)))

    for idx in rows_to_change:
        current_val = df.at[idx, column_name]
        
        if current_val in country_list:
            new_val = random.choice([c for c in country_list if c != current_val])
            df.at[idx, column_name] = new_val
    return df
glavne_drzave = ['Portugal', 'United Kingdom', 'France', 'Spain', 'Italy', 'Germany']
df = randomize_location(df, 'country', glavne_drzave)
```

```python
df = df.dropna()
print("CSV size after: ", df.shape)
print(df.head())
```

```python
df20 = df.sample(frac=0.2, random_state=1)
df = df.drop(df20.index)
print("CSV size 80: ", df.shape)
print("CSV size 20: ", df20.shape)
```
